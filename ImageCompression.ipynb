**Notation.** For any matrix $A$, by $A^t$ we mean transpose of $A$. Positve integer means an integer greater than or equal to $1$.

**Defintion.** A matrix $Q \in \mathbb{R}^{n \times n}$ is called an orthogonal matrix if the following holds

$$Q^t Q = Q^t Q = I.$$

**Theorem.** Let $A \in \mathbb{R}^{n \times n}$ be a  symmetric matrix. Then $A$ has real eigenvalues $\lambda_1, \dots, \lambda_n$ (with possible repetitions). Furthermore, $\mathbb{R}^n$ has a basis $v_1, \dots, v_n$ such that each $v_i$ is an eigenvector of $A$ with eigenvalue $\lambda_i$ and $A$ can be written as

$$A = Q D Q^t,$$

where $Q$ is an orthogonal matrix whose columns are $v_1, \dots, v_n$, and $D$ is the diagonal matrix whose diagonal entries are $\lambda_1, \dots, \lambda_n$.

A singular value decomposition (SVD) is a generalization of this fact where $A \in \mathbb{R}^{m \times n}$ is not necessarily symmetric or even square.

**Lemma.** Let $A \in \mathbb{R}^{m \times n }$. Now $A^tA$ is symmetric, so its eigenvalues are real. If $\lambda$ is an eignenvalue of $A^t A$, then $\lambda \geq 0$.  

Let $\lambda_1, \dots, \lambda_n$ denote the eigenvalues of $A^t A$ with possible repetitions. Order these so that $\lambda_1 \geq \lambda_2 \geq  \cdots \lambda_n \geq 0$. Let $\sigma_i = \sqrt{\lambda_i}$, so that $\sigma_1 \geq \sigma_2 \geq \cdots \geq \sigma_n \geq 0.$

**Definition.** The numbers $\sigma_1 \geq \cdots \geq \sigma_n \geq 0$ defined above are called the **singular values** of $A$.


**Proposition.** The number of nonzero singular values of $A$ equals the rank of $A$.

**Definition.** Let $A \in \mathbb{R}^{m \times n}$ be a matrix of rank $r$. Then a **singular value decomposition** of $A$ is a factorization

$$A = U \Sigma V^t,$$

where 

- $U$ is a real $m \times m$ orthogonal matrix

- $V$ is a real $n \times n$ orthogonal matrix

- $\Sigma$ is an $m \times n$ matrix whose $i$-th diagonal entry equals the $i$-th singular value $\sigma_i$ for $i = 1, \dots, r$. All other entries of $\Sigma$ are zero.

The columns of $U$ and the columns of $V$ are called the **left-singular vectors** and **right-singular vectors** of $A$, respectively. We can write $A$ as the sum of rank $1$ matrices

$$A = \sum_{i=1}^{r} \sigma_{i} u_i v_i^t,$$

where $u_i$ and $v_i$ are the $i$-th column of $U$ and $V$ respectively.

When $m \geq n$, the matrix $\Sigma$ has at most $n$ non-zero elements of the diagonal, and may be written as $\Sigma = \begin{bmatrix} \hat{\Sigma} \\ 0\end{bmatrix}$. Therefore, it is possible to exactly represent $A$ using the so-called **economy** SVD as oppossed to the **full** SVD:

$$A = U \Sigma V^t = \begin{bmatrix} \hat{U} & \hat{U}^{\perp} \end{bmatrix} \begin{bmatrix} \hat{\Sigma} \\ 0\end{bmatrix} V^t = \hat{U} \hat{\Sigma} V^t,$$

where $\perp$ denotes the orthogonal complement.

**Definition.** A matrix norm on the space $\mathbb{R}^{m \times n}$ is a function $f : \mathbb{R}^{m \times n}  \rightarrow \mathbb{R}$ which satisfies the following conditions for all $A, B \in \mathbb{R}^{m \times n}$ and $\lambda \in \mathbb{R}$,

- $f(A) \geq 0$ and $f(A) = 0$ if and only if $A = 0$.
- $f(A + B) \leq f(A) + f(B)$
- $f(\lambda A) = \lvert r \rvert f(A).$

**Examples.** 

- If $v = (v_1, \dots, v_n)^t \in \mathbb{R}^{n}$, then for a positive integer $p$, the $p$-norm of $v$ is defined as 

$$ \| v \|_p = (\lvert v_1 \rvert^{p} + \cdots + \lvert v_n \rvert^{p})^{1/p}.$$

- The infinity norm is defined as $\| v\|_{\infty} = \max \{ \lvert v_1 \rvert, \dots, \lvert v_n \rvert \}.$

- The Frobeinius norm of $A \in \mathbb{R}^{m \times n}$ is defined as 

$$\| A \|_{F} = \big(\sum_{i, j} a_{ij}^{2} \big)^{\frac{1}{2}}.$$ 

One can show that $\| A\|_{F} = \sqrt{\text{Trace}(AA^t)} = \sqrt{\sigma_1^2 + \dots + \sigma_r^2}.$


- Let $p$ be a positive integer. Then the spectral $p$-norm is defined as  $\| A \|_{p} = \sup_{x \neq 0} \frac{\|A x\|_p}{\| x\|_p}.$ 

$\| A \|_2$ is simply denoted by $\| A \|$.

## Best Rank $k$ Approximations

We proceed with the above settings. For $k \in \{ 1, \dots, r \}$, let 

$$A_k = \sum_{i=1}^{k} \sigma_i u_i v_i^t =U_k \Sigma_k V_k^t,$$

where $U_k$ and $V_k$ denote the first $k$ leading columns of $U$ and $V$, and $\Sigma_k$ contains the
leading $k \times k$ sub-block of $\Sigma$. $A_k$ has rank $k$. Furthermore, $A_k$ is the best rank $k$ approximation to $A$ when the error is measured in either the $2$-norm or the Frobenius norm. To be more precise:

**Theorem (Eckart–Young–Mirsky).** For any matrix $B \in \mathbb{R}^{m \times n}$ of rank at most $k$

$$ \sqrt{\sigma_{k+1}^2 + \cdots + \sigma_{m}^2}=\| A - A_k \|_F \leq \| A - B\|_F.$$

The same holds for $\| \cdot \|_2$, but there we have $\| A - A_k \|_2 = \sigma_{k+1} $.

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
plt.rcParams['figure.figsize'] = [14,8]
```

imread reads an image from a file into an array. This array has shape 
(M,N,3) for RGB images.

To convert RGB into grayscale we have to the take the average of three values, that is, (R+G+B)/3.

```python
A = imread('china.jpg')
X = np.mean(A, -1); #Convert RGB to grayscale
```

imshow displays the grayscale image in a figure.

```python
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices = False) # economy SVD
S = np.diag(S)
```

```python
j = 0
for r in (5,20,100): #ranks
    # Construct approximate image
    Xapprox = U[:, :r] @ S[0:r,:r] @ VT[:r,:] # @ means matrix multiplication
    plt.figure(j+1)
    j +=1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S)) # log(sigma_j) vs j
plt.title('Singular Values')
plt.show()

plt.figure(2)

# sum of the first j eigenvalues as fraction of 
# the sum of all eigenvalues
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S))) 

plt.title('Singular Values: Cumulative Sum')
plt.show()
```
